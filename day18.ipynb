#### Graph Functions
1. plot-Simple line plot
2. plot.area-Filled Version of Line Plot
3. plot.bar/plot.barh-Bar Graph
4. plot.box-Box Plot
5. plot.density-Density Line (Smooth Curved)
6. plot.hexbin-Hexagon Version of Scatter with Density 
7. plot.hist-Histogram
8. plot.kde-kernal Density Estimator (Density Graph)
9. plot.pie-Pie Chart of Categorical Data
10. plot.scatter-Scatter Graph

```python
df2.plot();plt.title('Line Plot'); plt.show()
df2.plot.area();plt.title('Area Plot'); plt.show()
df2.plot.bar();plt.title('Bar Plot'); plt.show()
df2.plot.barh();plt.title('Barh Plot'); plt.show()
df2.plot.box();plt.title('Box Plot'); plt.show()
df2.plot.density();plt.title('Density Plot'); plt.show()
df2.plot.kde();plt.title('Kde Plot'); plt.show()
df2.plot.hist();plt.title('Hist Plot'); plt.show()
df2.plot.scatter(x='a',y='b');plt.title('Scatter Plot'); plt.show()
df2.plot.hexbin(x='a', y='b',C='c', gridsize=6);plt.title('Hexbin Plot'); plt.show()
```

# Seaborn

```python
import seaborn
seaborn.set_style('whitegrid')
```

##### Categories
1. Categorical Plots (hue can be used | comparison with categorical data)
   1. countplot (single parameter)
   2. barplot (one categorical & one numerical | also calculate mean by default)
   3. boxplot
   4. violinplot
   5. swarmplot
2. Distribution Plots
   1. distplot
   2. displot
   3. jointplot
   4. pairplot
   5. kdeplot
   6. rugplot
   7. scatterplot
3. Matrix Plots
   1. heatmap
   2. clustermap
   3. pairplot (same as above one)
4. Regression Plots
   1. lmplot
   2. regplot
5. Grid Plots
   1. PairGrid
   2. JointGrid 
   3. FacetGrid

```python
seaborn.countplot(x='day', data=tips)
```

```python
seaborn.countplot(x='sex', data=tips)
```

```python
seaborn.countplot(x='time', data=tips)
```

```python
seaborn.countplot(x='smoker', data=tips , hue='sex')
```

```python
seaborn.countplot(x='day', data=tips , hue='sex')
```

```python
seaborn.barplot(x='day', y='total_bill',data=tips)
```

```python
seaborn.barplot(x='day', y='total_bill',data=tips, estimator=sum)
```

```python
seaborn.barplot(x='day', y='total_bill',data=tips, hue='time', estimator=np.mean)
```

```python
seaborn.boxplot(x='day', y='total_bill',data=tips , hue='smoker')
```

```python
seaborn.swarmplot(x='day', y='total_bill', data=tips , hue='smoker')
```

```python
seaborn.violinplot(x='day', y='total_bill', data=tips , hue='smoker')
```

```python
tips.head()
```

```python
seaborn.distplot(tips['total_bill'])
```

```python
seaborn.displot(data=tips, x='total_bill')
```

```python
seaborn.displot(data=tips, x='total_bill', y='tip')
```

```python
seaborn.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
seaborn.jointplot(data=tips, x='total_bill', y='tip',hue='smoker')
plt.show()
seaborn.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
plt.show()

seaborn.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
fig = plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
seaborn.kdeplot(data=tips, x='total_bill')
seaborn.rugplot(data=tips, x='total_bill')
plt.subplot(1,2,2)
seaborn.kdeplot(data=tips, x='total_bill', y='tip')
seaborn.rugplot(data=tips, x='total_bill', y='tip')
```

```python
iris = seaborn.load_dataset('iris')
```

```python
iris.head()
```

```python
seaborn.pairplot(data= iris, hue= 'species')
```

```python
seaborn.regplot(data=tips,x='total_bill',y='tip')
plt.show()
seaborn.lmplot(data=tips, x='total_bill', y='tip')
```

```python
pg = seaborn.PairGrid(data=iris)
pg.map_diag(seaborn.kdeplot)
pg.map_lower(seaborn.scatterplot)
pg.map_upper(seaborn.kdeplot)
```

```python
jg = seaborn.JointGrid(data=tips, x='total_bill', y='tip')
jg.plot(seaborn.regplot, seaborn.distplot)
```

```python
fg = seaborn.FacetGrid(data=tips, row='smoker', col='day', hue='sex')
fg.map(seaborn.scatterplot, 'total_bill', 'tip').add_legend()
```
