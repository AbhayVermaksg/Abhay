```python
for col in cat_cols:
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        seaborn.boxplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,2)
        seaborn.violinplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.subplot(1,3,3)
        seaborn.barplot(data=data,x=col,y=ncol,palette=choice(p_list))
        plt.show()
```

```python
for col1 in num_cols:
    for col2 in num_cols:
        if col1 == col2:
            continue
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col1} vs {col2}')
        plt.subplot(1,2,1)
        seaborn.jointplot(data=data,x=col1,y=col2,palette=choice(p_list))
        plt.subplot(1,2,2)
        seaborn.lineplot(data=data,x=col1,y=col2,palette=choice(p_list))
        plt.show()
```

```python
pg = seaborn.PairGrid(data= data[num_cols], palette=choice(p_list))
pg.map_diag(plt.hist)
pg.map_lower(seaborn.regplot)
pg.map_upper(seaborn.kdeplot)
```

```python
seaborn.heatmap(data= data[num_cols].corr(),annot = True, cmap= choice(p_list))
```

# Multuivariate Analysis

```python
for col in cat_cols:
    if col=='Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        seaborn.boxplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,2)
        seaborn.violinplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.subplot(1,3,3)
        seaborn.barplot(data=data,x=col,y=ncol,hue='Product',palette=choice(p_list))
        plt.show()
```
