```python
import matplotlib.pyplot as plt
import numpy as np
```

# Types of Graphs
1. Plot        - line graph , between two parameters
2. Bar         - bar graph of single parameter
3. Hist        - histogram of a bunch of data i.e. frequency of single parameter
4. Scatter     - dot graph between two parameters
5. Boxplot/Box - statstical overview in form of box for a single parameter
                 mean, outliners, quantiles (25%, 75%), IQR
                 (optional - Categorical data)
6. Stem        - more like bar graph, Lines instead of bars 
7. Violinplot  - Violen Shaped Statistical overview graph
                 spread of data(more like hist)
                 single parameter (optional cat-data)

```python
x = np.linspace(0,5,10)
x1 = np.arange(1,11,3)
```

```python
x
```

```python
y=x**3
```

```python
y2=x**4
```

```python
fig= plt.figure(figsize= (10,7))
plt.plot(x,y, lw=3, marker='o', markersize=15, markerfacecolor='yellow')
plt.plot(x,y2, marker = 'v', markersize = 10)
plt.legend(['y=x**3','y=x**4'])
plt.grid()
plt.show()
```

```python
fig.savefig('mygraph.png')
```

```python
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()
```

```python
plt.hist(x)
plt.title('Histogram')
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Scatter')
plt.grid()
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(x)
plt.title('Box Plot')
plt.show()
```

```python
plt.stem(x,y)
plt.title('Stem')
plt.show()
```

```python
plt.violinplot(x)
plt.title('Violen Plot - Badsoorat wala')
plt.show()
```
